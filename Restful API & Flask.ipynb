```python
#Q 1: What is a RESTful API?

Answer:_A RESTful API (Representational State Transfer API) is a type of web service that allows communication
        between systems over HTTP. It follows REST principles such as statelessness, client-server architecture,
        and use of standard HTTP methods.In a REST API, resources (like users, products) are identified by URLs, 
        and operations are performed using methods like GET, POST, PUT, DELETE. Data is usually exchanged in JSON format.
            
#Q2: What is Flask, and why is it popular for building APIs?

Answer:Flask is a lightweight web framework written in Python used to build web applications and APIs.
       It is popular because:
       It is simple and easy to learn
       Minimal setup (micro-framework)
       Flexible and customizable
       Supports extensions (like database, authentication)
       Ideal for building REST APIs quickly

#Q 3: What are HTTP methods used in RESTful APIs?

Answer:HTTP methods define the type of operation to be performed:
       GET → Retrieve data
       POST → Create new data
       PUT → Update existing data
       PATCH → Partially update data
       DELETE → Remove data

#Q 4: What is the purpose of the @app.route() decorator in Flask?

Answer:The @app.route() decorator is used to map a URL to a Python function.
        When a user visits a specific URL, Flask executes the associated function and returns the response
```

```python
@app.route('/home')
def home():
    return "Welcome"
```

```python
#Q 5: What is the role of Flask-SQLAlchemy?

Answer:Flask-SQLAlchemy is an extension that integrates SQLAlchemy (ORM) with Flask.
       It helps in:
       Interacting with databases using Python classes
       Avoiding raw SQL queries
       Managing database models and relationships
       Simplifying database operations (CRUD)
    
#Q 6: How do you create a basic Flask application?

Answer:
       Steps:
       Install Flask
       Create a Python file
       Import Flask and create an app instance
       Define routes
       Run the app
    from flask import Flask
app = Flask(name)

@app.route('/')
def home():
    return "Hello World"

if name == 'main':
    app.run(debug=True)
```

```python
#Q 7: How do you return JSON responses in Flask?

Answer:Use jsonify() function to return JSON data.
Example:
from flask import jsonify

@app.route('/data')
def data():
    return jsonify({"name": "Sudhanshu", "age": 22})
```

```python
#Q 8: How do you handle POST requests in Flask?

Answer:
Use methods=['POST'] in route and access data using request.
Example:
from flask import request

@app.route('/submit', methods=['POST'])
def submit():
    data = request.json
    return {"received": data}
    
```

```python
#Q 10: How do you structure a Flask app using Blueprints?
Answer:
Blueprints help organize large applications into modules.
Steps:
Create a blueprint
Define routes inside it
Register it in the main app
Example:
from flask import Blueprint

main = Blueprint('main', name)

@main.route('/')
def home():
    return "Home Page" 
```
